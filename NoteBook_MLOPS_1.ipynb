```python
# Importation des packages
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# CrÃ©ation du DataFrame Data (vide) qui contiendra le chiffre d'affaire par heure
data = pd.DataFrame(columns = ["DateTime", "Sales_Revenue"])
```

```python
# CrÃ©ation d'une liste avec le nom de chaque fichier CSV contenue dans le dossier CSV
path = "./CSV"
files = [file for file in os.listdir(path) if file.endswith(".csv")]
```

```python
data_list = [] # CrÃ©ation d'une liste vide qui contiendra les DataFrame regroupant le chiffre d'affaire par heure de chaque fichier
# RÃ©cupÃ©ration du chiffre d'affaire par heure pour l'ensemble des fichiers CSV
for file in files:
    df = pd.read_csv(os.path.join(path, file)) # Import du fichier
    df["Order Date"] = pd.to_datetime(df["Order Date"]) # Conversion de la variable Order Date au format date
    df["DateTime"] = df["Order Date"].dt.floor("h") # Arrondi la Order Date Ã  l'heure Ã  infÃ©rieure la plus proche
    df["Sales_Revenue"] = df["Quantity"] * df["Product Price"] 
    data_list.append(df) # Ajout du DataFrame contenant le chiffre d'affaire par heure du fichier Ã  la liste 
    print(file)
data = pd.concat(data_list, ignore_index = True) # CrÃ©ation du fichier final regroupant l'ensemble des DataFrame de la liste
data = data.groupby("DateTime", as_index = False)["Sales_Revenue"].sum()
```

```python
time_freq = pd.Series(pd.date_range(start=data["DateTime"].min(), end=data["DateTime"].max(), freq="h"))
time_freq_df = pd.DataFrame(time_freq, columns=['DateTime'])
```

```python
merged_data = pd.merge(time_freq_df, data, on="DateTime", how="left")
merged_data.loc[merged_data['Sales_Revenue'].isna(), 'Sales_Revenue'] = 0
```

```python
# CrÃ©ation des variables Year, Month, Day et Hour
merged_data["Year"] = merged_data["DateTime"].dt.year
merged_data["Month"] = merged_data["DateTime"].dt.month
merged_data["Day"] = merged_data["DateTime"].dt.day_name()
merged_data["Hour"] = merged_data["DateTime"].dt.hour
```

```python
# Trie de la sortie par date
merged_data = merged_data.sort_values(by = "DateTime").reset_index(drop = True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(merged_data["DateTime"], merged_data["Sales_Revenue"], c = "lightblue")
%matplotlib
```
